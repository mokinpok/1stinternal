```python
cars_df=pd.read_csv('C:\\Users\\GPT BANTWAL\\Downloads\\mtcars (1).csv')
cars_df
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
range=(10,40)
bins=10
plt.hist(cars_df['mpg'],bins,range,color='red',rwidth=0.8)
plt.xlabel("mpg")
plt.ylabel("No of frequency")
plt.title("Histogram")
plt.show()
```

```python
plt.scatter(cars_df['wt'],cars_df['mpg'],color="red",marker="*")
plt.xlabel("Weight of thr cars")
plt.ylabel("mpg")
plt.title("Scatter plot")
plt.show()                  
```

```python
temp=cars_df['am'].value_counts()
print(temp)
temp.plot.bar()
plt.xlabel("Transmission type")
plt.ylabel("count")
plt.title("bar graph")
plt.show()
```
